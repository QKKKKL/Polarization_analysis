```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_kde_from_file(file_path, phase_name):

    data = pd.read_csv(file_path)

    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle(f'{phase_name} Plots', fontsize=16)
    
    # Amplitude 
    sns.kdeplot(data=data, x='amplitude_val', weights=data['amplitude_freq'], ax=axes[0], fill=True)
    axes[0].set_title(f'{phase_name} Amplitude')
    axes[0].set_xlabel('Amplitude')
    axes[0].set_ylabel('Frequency')
    
    # Azimuth 
    sns.kdeplot(data=data, x='azimuth_val', weights=data['azimuth_freq'], ax=axes[1], fill=True)
    axes[1].set_title(f'{phase_name} Azimuth')
    axes[1].set_xlabel('Azimuth')
    axes[1].set_ylabel('Frequency')
    
    # Inclination 
    sns.kdeplot(data=data, x='inclination_val', weights=data['inclination_freq'], ax=axes[2], fill=True)
    axes[2].set_title(f'{phase_name} Inclination')
    axes[2].set_xlabel('Inclination')
    axes[2].set_ylabel('Frequency')
    
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
```

```python
plot_kde_from_file("kde_df_data/IU.HRV.1_kde_p.csv", "IU HRV 1 P Wave")
plot_kde_from_file("kde_df_data/IU.HRV.1_kde_s.csv", "IU HRV 1 S Wave")
plot_kde_from_file("kde_df_data/IU.HRV.1_kde_noise.csv", "IU HRV 1 Noise")
```

```python
def plot_kde_from_file(file_path, phase_name, save_directory):

    data = pd.read_csv(file_path)

    # Extract event name (the part before first _)
    event_name = os.path.basename(file_path).split('_')[0]

    # Create figure with 3 subplots
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle(f'Event: {event_name} - {phase_name} Plots', fontsize=16)

    # Amplitude 
    sns.kdeplot(data=data, x='amplitude_val', weights=data['amplitude_freq'], ax=axes[0], fill=True)
    axes[0].set_title(f'{phase_name} Amplitude')
    axes[0].set_xlabel('Amplitude')
    axes[0].set_ylabel('Frequency')

    # Azimuth
    sns.kdeplot(data=data, x='azimuth_val', weights=data['azimuth_freq'], ax=axes[1], fill=True)
    axes[1].set_title(f'{phase_name} Azimuth')
    axes[1].set_xlabel('Azimuth')
    axes[1].set_ylabel('Frequency')

    # Inclination 
    sns.kdeplot(data=data, x='inclination_val', weights=data['inclination_freq'], ax=axes[2], fill=True)
    axes[2].set_title(f'{phase_name} Inclination')
    axes[2].set_xlabel('Inclination')
    axes[2].set_ylabel('Frequency')

    plt.tight_layout(rect=[0, 0, 1, 0.95])

    # Save plot to the specified directory
    file_name = os.path.basename(file_path).replace('.csv', '.png')
    plot_save_path = os.path.join(save_directory, file_name)
    plt.savefig(plot_save_path)
    plt.close(fig)

    print(f'Saved plot: {plot_save_path}')
```

```python
def process_all_kde_files(input_directory, save_directory):

    # if save directory does not exist, create one
    os.makedirs(save_directory, exist_ok=True)

    for file_name in os.listdir(input_directory):
        file_path = os.path.join(input_directory, file_name)
        if '_kde_p' in file_name:
            phase_name = 'P'
        elif '_kde_s' in file_name:
            phase_name = 'S'
        elif '_kde_noise' in file_name:
            phase_name = 'Noise'
        else:
            continue

        # plot and save 
        plot_kde_from_file(file_path, phase_name, save_directory)


input_directory = './kde_df_data'  
save_directory = './kde_plots' 

process_all_kde_files(input_directory, save_directory)
```
